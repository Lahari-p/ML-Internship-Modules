### 6.2 Calculating average annual percentage return and volatality for one year.

```python
avg_annual_return = (final_df.pct_change().mean()) * 252
avg_n_vol_df = pd.DataFrame(avg_annual_return)
avg_n_vol_df.columns = ['annual returns']
avg_n_vol_df['volatality'] = (final_df.pct_change().std()) * np.sqrt(252)
avg_n_vol_df
```

### 6.3 Clustering 30 stocks using KMeans  and finding optimal k using Elbow curve method.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(avg_n_vol_df)
norm_data = sc.transform(avg_n_vol_df)
k=range(1,10)
sum_of_squares_dist = []
for i in k:
    km = KMeans(n_clusters = i)
    km = km.fit(norm_data)
    sum_of_squares_dist.append(km.inertia_)
plt.figure(figsize=(12,6))
plt.plot(k,sum_of_squares_dist,'bx-')
plt.show()
```

### 6.4 Preparing seperate dataframe to show which stocks belong to same cluster.

```python
km1 = KMeans(n_clusters = 5)
km1.fit(norm_data)
```

```python
clustered_df = pd.DataFrame(data = avg_n_vol_df)
clustered_df['labels'] = km1.labels_
clustered_df
```

```python
plt.figure(figsize=(16,8))
x = clustered_df['annual returns'].values
y = clustered_df['volatality'].values
types = clustered_df.reset_index()['index'].values

plt.scatter(clustered_df['annual returns'][clustered_df.labels==0],clustered_df.volatality[clustered_df.labels==0],s=100,color='g',marker='o',label='cluster1')
plt.scatter(clustered_df['annual returns'][clustered_df.labels==1],clustered_df.volatality[clustered_df.labels==1],s=100,color='r',marker='o',label='cluster2')
plt.scatter(clustered_df['annual returns'][clustered_df.labels==2],clustered_df.volatality[clustered_df.labels==2],s=100,color='b',marker='o',label='cluster3')
plt.scatter(clustered_df['annual returns'][clustered_df.labels==3],clustered_df.volatality[clustered_df.labels==3],s=100,color='k',marker='o',label='cluster4')
plt.scatter(clustered_df['annual returns'][clustered_df.labels==4],clustered_df.volatality[clustered_df.labels==4],s=100,color='y',marker='o',label='cluster5')

for i, txt in enumerate(types):
    plt.annotate(txt, (x[i], y[i]), xytext=(10,10), textcoords='offset points')
plt.legend(loc='upper right')
plt.show()
```
